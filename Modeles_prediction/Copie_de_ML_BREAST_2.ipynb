```python
# Reading Dataset:
dataset = pd.read_csv("/content/drive/MyDrive/PROJET 3/INFO MALADIE/DF/df_nettoyer/dfcancersein.csv")
# Top 5 records:
dataset.describe()
```

```python
# Splitting Independent and Dependent Feature:
X_rfcs = dataset[['radius_mean','texture_mean','smoothness_mean','compactness_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','smoothness_se','compactness_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','smoothness_worst','compactness_worst','symmetry_worst','fractal_dimension_worst']]
y_rfcs = dataset.iloc[:, 0]
```

```python
X_rfcs.describe()
```

```python
dataset['diagnosis'].value_counts()
```

***RANDOM FOREST***

```python
# Train Test Split:
from sklearn.model_selection import train_test_split
X_rfcs_train,X_rfcs_test,y_rfcs_train,y_rfcs_test = train_test_split(X_rfcs,y_rfcs, test_size=0.3, random_state=42)
```

```python
print(X_rfcs_train.shape)
print(X_rfcs_test.shape)
```

```python
# Importing Performance Metrics:
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# RandomForestClassifier:
from sklearn.ensemble import RandomForestClassifier
random_forest_model_rfcs = RandomForestClassifier()
random_forest_model_rfcs.fit(X_rfcs_train,y_rfcs_train)

# Predictions:
y_pred = random_forest_model_rfcs.predict(X_rfcs_test)

# Performance:
print('Accuracy:', accuracy_score(y_rfcs_test,y_pred))
print(confusion_matrix(y_rfcs_test,y_pred))
print(classification_report(y_rfcs_test,y_pred))
```

```python
train_score_rfcs = random_forest_model_rfcs.score(X_rfcs_train, y_rfcs_train)
test_score_rfcs = random_forest_model_rfcs.score(X_rfcs_test, y_rfcs_test)
print(f"- Le score pour le jeu d'entrainement est : {train_score_rfcs*100:.2f}%")
print(f"- Le score pour le jeu de test est : {test_score_rfcs*100:.2f}%")
```

***LOGISTIC REGRESSION***

```python
X_lrcs = dataset[['radius_mean','texture_mean','smoothness_mean','compactness_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','smoothness_se','compactness_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','smoothness_worst','compactness_worst','symmetry_worst','fractal_dimension_worst']]
y_lrcs = dataset.iloc[:, 0]
```
